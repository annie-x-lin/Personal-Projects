```python
# Define host
import socket
import threading
from tkinter import *
import datetime
client1 = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

host = 'your servers local ip address'
host = socket.gethostbyname(socket.gethostname()) #replace with your server's local ip address


# define the communication port
port = 5002
server_address = ((host, port))
print("Connecting...\n")
client1.connect(server_address)
print("Connected!")

class gui:
    def __init__(self):
        self.window=Tk()
        self.window.iconbitmap(r'C:\Users\annie\Desktop\Homework\EENG202\Personal Projects\Chat.ico')

        self.window.withdraw()

        self.login=Toplevel()
        self.login.iconbitmap(r'C:\Users\annie\Desktop\Homework\EENG202\Personal Projects\Chat.ico')
        self.login.resizable(width = False,height = False)
        self.login.geometry("300x140")

        self.loginlab=Label(self.login, text="Login to continue", justify="center")
        self.loginlab.pack()

        self.user=Label(self.login, text="Username:  ")
        self.user.place(rely = 0.2,relx=0.1)
        
        self.enterName = Entry(self.login)
        self.enterName.place(rely = 0.2,relx=0.5)
        
        self.passlab=Label(self.login, text="Password: ")
        self.passlab.place(rely = 0.4,relx=0.1)
        
        self.enterPass = Entry(self.login)
        self.enterPass.place(rely = 0.4,relx=0.5)
        
        self.Sub = Button(self.login, text="Submit")
        self.Sub.config(command=lambda: self.cont(self.enterName.get(),self.enterPass.get()))
        self.Sub.place(rely = 0.7, relx = 0.4) 
        self.login.bind('<Return>',lambda event:self.cont(self.enterName.get(),self.enterPass.get()))
        
        self.window.mainloop()

    def cont(self, name, password):
        self.layout(name, password)
        rcv = threading.Thread(target=self.receive)
        rcv.start()
        
        self.login.withdraw()

    def sendMessage(self):
        self.textbox.config(state=DISABLED)
        while True:
            message = str(self.name)+": "+str(self.msg)
            client1.send(message.encode('utf-8'))   
            break 

    def Send(self, message, name):
        self.msg=message
        snd= threading.Thread(target = self.sendMessage)
        snd.start()
        self.EnterMsg.delete(0, END)
        
    def quit(self,name):
        client1.send((name+" has left the chat.").encode('utf-8'))
        client1.shutdown(2)
        self.window.destroy()
        
    def receive(self):
        while True:
            try:
                message = client1.recv(500).decode('utf-8')
                # if the messages from the server is NAME send the client's name
                if message == 'NAME':
                    client1.send(self.name.encode('utf-8'))
                elif message =="PASSWORD":
                    client1.send(self.password.encode('utf-8'))
                else:
                    # insert messages to text box
                    self.textbox.config(state = NORMAL)
                    self.textbox.insert(END,message+"\n\n")
                    self.textbox.config(state = DISABLED)
                    self.textbox.see(END)
            except:
                # an error will be printed on the command line or console if there's an error
                print("An error occured!")
                self.textbox.config(state = NORMAL)
                self.textbox.insert(END,"WRONG PASSWORD/USER, TRY AGAIN")
                self.textbox.config(state = DISABLED)
                self.textbox.see(END)
                break
                
    

    def layout(self, name, password):
        self.name=name
        self.password=password
        self.window.deiconify()
        self.window.title("Chatbox")
        self.window.geometry("430x530")
        self.window.configure(bg = "#0DA2FF")
        self.window.resizable(width = True,height = True)
        
        self.nameLabel = Label(self.window,bg = "#394d6d", fg = "#FFFFFF",text = "ChatRoom", font = ("Cambria", 14, "bold"), pady = 5,width = 5,relief="ridge")
        self.nameLabel.place(relheight = 1/12,relwidth = 1)
        
        self.photo= PhotoImage(file="person.png")
        self.profile = Label(self.nameLabel, image=self.photo, text = "  "+name, compound = "left" ,font = ("Cambria", 11, "bold"),fg = "#FFFFFF", bg= "#394d6d")
        self.profile.place(relx = 0.03, rely = 0.15)
        

        self.exit = Button(self.nameLabel, text="X", command = lambda : self.quit(name),bg = "#23395d", fg = "#FFFFFF", highlightthickness = 0, bd = 0)
        self.exit.place(relx = 0.92, rely = 0.35, relheight = 0.37, relwidth = 0.035)
        
        self.textbox = Text(self.window,width = 20, height = 2,bg = "#141E2F",fg = "#FFFFFF",highlightthickness = 0, bd = 0)
        self.textbox.place(relheight = 0.77,relwidth = 1,rely = 0.08)
        self.textbox.insert("end","Welcome to your chatroom!\n\n"+str(datetime.datetime.now())+"\n")
        
        self.labelBottom = Label(self.window,bg = "#394d6d", height = 80,highlightthickness = 0, bd = 0)
        self.labelBottom.place(relwidth = 1, rely = 0.85)

        self.EnterMsg = Entry(self.labelBottom,bg = "#141E2F", fg = "#FFFFFF",highlightthickness = 0, bd = 0)
        self.EnterMsg.place(relwidth = 0.74,relheight = 0.05,rely = 0.008,relx = 0.015)
        
        self.sendimage = PhotoImage(file="send.png")
        self.send = Button(self.labelBottom, image=self.sendimage,bg = "#394d6d",highlightthickness = 0, bd = 0, command=lambda : self.Send(self.EnterMsg.get(),name), width=20)
        self.send.place(relx = 0.77, rely = 0.008, relheight = 0.05, relwidth = 0.18)
        self.window.bind('<Return>',lambda event:self.Send(self.EnterMsg.get(),name))
        
        self.scrollbar = Scrollbar(self.textbox)    
        self.scrollbar.place(relheight = 1, relx = 0.974)
        self.scrollbar.config(command = self.textbox.yview)        

g= gui()
```
