```python
import random
list1=["R","P","S"]

def rps(a):
    print("           Welcome to Rock Paper Scissors!")
    print("")
    print("Rules: type in R for rock, P for paper, and S for scissor")
    print("")
    playercount=0
    compcount=0
    count=0
    while count<5:
        print ("R, P, or S?")
        var1=input("Player: ")
        var2 = random.choice(a)
        count+=1
        if var1 == "R":
            if var2 == "R":
                print ("Computer: R")
                print ("draw")
                print("")
            elif var2 == "P":
                print("Computer: P")
                compcount +=1
                print ("Computer +1")
                print("")
            else:
                print("Computer: S")
                playercount+=1
                print ("Player +1")
                print("")
        elif var1 == "P":
            if var2 == "P":
                print ("Computer: P")
                print ("draw")
                print("")
            elif var2 == "S":
                print("Computer: S")
                compcount +=1
                print ("Computer +1")
                print("")
            else:
                print("Computer: R")
                playercount+=1
                print ("Player +1")
                print("")
        elif var1 == "S":
            if var2 == "S":
                print ("Computer: S")
                print ("draw")
                print("")
            elif var2 == "R":
                print("Computer: R")
                compcount +=1
                print ("Computer +1")
                print("")
            else:
                print("Computer: P")
                playercount+=1
                print ("Player +1")
                print("")
    if playercount == compcount:
        print("DRAW")
    elif playercount > compcount:
        print ("YOU WIN")
    else:
        print("YOU LOST")
rps(list1)
```
